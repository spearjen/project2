```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st
import numpy as np
import zipfile as zf

# Study data files
hurricane_data = pd.read_csv("C:/Users/spearjen/Desktop/unzipped/atlantic.csv")
df1851_2000=pd.read_csv("C:/Users/spearjen/Desktop/unzipped/df1851_2000.csv")
df2001_2004= pd.read_csv("C:/Users/spearjen/Desktop/unzipped/df2001_2004.csv")
df2005=pd.read_csv("C:/Users/spearjen/Desktop/unzipped/df2005.csv")
df2006=pd.read_csv("C:/Users/spearjen/Desktop/unzipped/df2006.csv")
df2007=pd.read_csv("C:/Users/spearjen/Desktop/unzipped/df2007.csv")
df2008=pd.read_csv("C:/Users/spearjen/Desktop/unzipped/df2008.csv")
df2009=pd.read_csv("C:/Users/spearjen/Desktop/unzipped/df2009.csv")
df2010=pd.read_csv("C:/Users/spearjen/Desktop/unzipped/df2010.csv")
df2011=pd.read_csv("C:/Users/spearjen/Desktop/unzipped/df2011.csv")
df2012=pd.read_csv("C:/Users/spearjen/Desktop/unzipped/df2012.csv")
df2013=pd.read_csv("C:/Users/spearjen/Desktop/unzipped/df2013.csv")
df2014=pd.read_csv("C:/Users/spearjen/Desktop/unzipped/df2014.csv")
df2015=pd.read_csv("C:/Users/spearjen/Desktop/unzipped/df2015.csv")
df2016=pd.read_csv("C:/Users/spearjen/Desktop/unzipped/df2016.csv")
df2017=pd.read_csv("C:/Users/spearjen/Desktop/unzipped/df2017.csv")
sst_at_data = pd.concat([df1851_2000, df2001_2004, df2005, df2006, df2007, df2008, df2009, df2010, df2011, df2012, df2013, df2014, df2015,
           df2016, df2017])
```

```python
hurricane_data.head()
```

```python
#filter to just landfall
hurricane_landfall = hurricane_data[(hurricane_data["Event"] == " L")]

    
index = hurricane_landfall.index
number_of_rows = len(index)
number_of_rows
```

```python
#format date
hurricane_landfall['Date'] = pd.to_datetime(hurricane_landfall['Date'], format='%Y%m%d')
hurricane_landfall.head()
```

```python
hurricane_landfall['Year'] = pd.DatetimeIndex(hurricane_landfall['Date']).year
hurricane_landfall['Month'] = pd.DatetimeIndex(hurricane_landfall['Date']).month
hurricane_landfall['Day'] = pd.DatetimeIndex(hurricane_landfall['Date']).day
hurricane_landfall.head()
```

```python
#Groupby year count
gb_hurricane_landfall = hurricane_landfall.groupby("Year").agg({'Event':['count']})
gb_hurricane_landfall
```

```python
gb_hurricane_landfall.plot.line(figsize=(5,5), color = "blue")
plt.legend("")
plt.title("Hurricane Landfall Count by Year 1851-2019")
plt.savefig("landfall_count_by_year")
```

```python
#Groupby year median windspeed
gb_hurricane_windspeed = hurricane_landfall.groupby("Year").agg({'Maximum Wind':['median']})
gb_hurricane_windspeed
```

```python
gb_hurricane_windspeed.plot.line(figsize=(5,5), color = "blue")
plt.legend("")
plt.title("Median Maximum Windspeed at Landfall by Year 1851-2019")
plt.savefig("windspeed_med_by_year")
```

```python
sst_at_data.head()
sst_at_data_drop=sst_at_data.dropna()
sst_at_data_drop
```

```python
#Groupby year median air temp 
gb_air_temp = sst_at_data_drop.groupby(["year","latitude","longitude"]).agg({'air_temperature':['median']})
gb_air_temp
```

```python
#Groupby year median sst
gb_sst = sst_at_data_drop.groupby(["year","latitude","longitude"]).agg({'sea_surface_temp':['median']})
gb_sst
```

```python
gb_air_temp.plot.line(figsize=(5,5), color = "blue")
plt.legend("")
plt.title("Air Temperature (Atlantic) by Year 2001-2004")
plt.savefig("air_temp_med_by_year")
```

```python
gb_sst.plot.line(figsize=(5,5), color = "blue")
plt.legend("")
plt.title("Sea Surface Temperature (Atlantic) by Year 2001-2004")
plt.savefig("sst_med_by_year")
```
